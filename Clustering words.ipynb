```python
import pandas as pd
df = pd.read_csv('data_from_news.csv', quotechar='"', skipinitialspace=True)

df
```

```python
df_cleaned = df.dropna(subset=['article_keywords'])
df_cleaned
```

```python
all_words = (df['article_keywords']) #.head(10))
all_words
```

```python
#pip install -U scikit-learn
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
from transformers import AutoTokenizer, AutoModel
import torch
```

```python
import numpy as np
```

```python
model_name = "sentence-transformers/all-MiniLM-L6-v2"  # Lightweight embedding model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

# Generating embeddings for all words
def get_embedding(word):
    inputs = tokenizer(word, return_tensors="pt", truncation=True, padding=True)
    outputs = model(**inputs)
    embedding = outputs.last_hidden_state.mean(dim=1).detach().numpy()
    return embedding.flatten()

embeddings = np.array([get_embedding(word) for word in all_words])
```

```python
embeddings
```

```python
from sklearn.cluster import KMeans

# Perform K-Means clustering
n_clusters = 5  # Number of clusters
kmeans = KMeans(n_clusters=n_clusters, random_state=0).fit(embeddings)

# Add cluster assignments to words
clusters = {word: cluster for word, cluster in zip(all_words, kmeans.labels_)}

# View clusters
for cluster_id in range(n_clusters):
    print(f"Cluster {cluster_id}: {[word for word, label in clusters.items() if label == cluster_id]}")
```

```python
from sklearn.manifold import TSNE

# Reduce embeddings to 2D
tsne = TSNE(n_components=2, random_state=0)
embeddings_2d = tsne.fit_transform(embeddings)

# Plot clusters
plt.scatter(embeddings_2d[:, 0], embeddings_2d[:, 1], c=kmeans.labels_, cmap='viridis')
plt.title("Word Clusters")
plt.xlabel("Dimension 1")
plt.ylabel("Dimension 2")
plt.show()
```
